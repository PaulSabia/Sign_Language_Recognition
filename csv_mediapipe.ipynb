# Ajout des coordonn√©es de la main droite pour chaque image dans le csv

```python
import mediapipe as mp
import cv2
import os
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
from time import sleep
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

## Test Mediapipe on Video

```python
cap = cv2.VideoCapture(0)

#Initiate holistic model:
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        #Recolor Feed
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        #Make detection
        result = holistic.process(img)
        
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        
        #Right hand
        mp_drawing.draw_landmarks(img, result.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        
        #Left hand
        mp_drawing.draw_landmarks(img, result.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        
        cv2.imshow('Test Mediapipe', img)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
cap.release()
cv2.destroyAllWindows()
```

# Test Mediapipe on Image

```python
path_images = [os.getcwd()+"\\dataset_final\\A\\A_027.jpg"]
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    for path_image in path_images:
        image_test = cv2.imread(path_image)
        image = cv2.cvtColor(image_test, cv2.COLOR_BGR2RGB)
        result = holistic.process(image)

        coords = result.right_hand_landmarks.landmark
        
        for y, coord in enumerate(coords):
            print(y ,' --- ', coord.x, coord.y, coord.z)

        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        #Right hand
        mp_drawing.draw_landmarks(image, result.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

        cv2.imshow('Image de Test',image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Generate CSV

```python
df = pd.read_csv('data.csv')

df.head()
```

```python
images_path = df['Path'].values

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    for i, row in df.iterrows():
        path = row.Path
        print(path)
        image = cv2.imread(path)
        img =  cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        sleep(0.5)
        results = holistic.process(img)
        try:
            coords = results.right_hand_landmarks.landmark
        
            for y, coord in enumerate(coords):
                #print(y ,' --- ', coord.x, coord.y, coord.z)
                df.at[i,f'x{y}'] = coord.x
                df.at[i,f'y{y}'] = coord.y
                df.at[i,f'z{y}'] = coord.z
            
        except Exception as e:
            print(e)
                            
```
